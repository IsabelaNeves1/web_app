# Data Analitycs - Produção de grãos estimada ciclo 2025/2026

```python
import pandas as pd
```

```python
caminho_drive = '/content/drive/MyDrive/analise_safra_conab_real.csv'
```

```python
df = pd.read_csv(caminho_drive, sep=';', encoding='utf-8-sig')
```

```python
print("--- Tabela Geral da Safra 2025/2026 ---")
display(df)
```

```python
# Somando a produção de todas as culturas da tabela
total_producao = df['Produção Estimada (Milhões t)'].sum()

# Calculando a porcentagem de representatividade de cada uma
df['Participação (%)'] = (df['Produção Estimada (Milhões t)'] / total_producao) * 100

display(df[['Cultura', 'Produção Estimada (Milhões t)', 'Participação (%)']])
```

```python
# Culturas com maior crescimento
vencedores = df.sort_values(by='Variação vs Safra Passada (%)', ascending=False)
print("--- Destaques Positivos e Negativos ---")
display(vencedores[['Cultura', 'Variação vs Safra Passada (%)']])
```
